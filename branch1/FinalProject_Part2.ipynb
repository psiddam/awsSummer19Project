```python
import numpy
import os
import re
import altair
import pandas as pd
altair.renderers.enable('notebook')
altair.data_transformers.disable_max_rows()
import matplotlib.pyplot as plt
```

```python
carriers = pd.read_csv("304914294_T_ONTIME_REPORTING.csv") 
```

```python
print(carriers.head())
```

```python
def calcStats(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}

```

```python
allStats = carriers['DEP_DELAY'].groupby(carriers['OP_CARRIER_AIRLINE_ID']).apply(calcStats).unstack()
allStats = allStats.sort_values('count')
allStats
```
